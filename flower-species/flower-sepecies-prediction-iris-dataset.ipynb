# Train test Split

# Prepare our data for modeling
This involves splitting the data back out into Numpy arrays

```python
# dropping the last 2 columns and return to new object instead of changing the original object
x_train= df_train.drop(columns=["target","target_names"]).values
y_train= df_train["target"].values
```

```python
print(np.array(y_train))
```

# Modeling - Baseline
what is the simplest model we can think of?
In this case, if our baseline model is just  randomly guessing flower species or guessing a single species for every data point, e would expect to have a model accuracy of 33%, since we have three different classes that are evenly balanced (50 data-points from each class)

so our model should have at least 0.33 or 33% accuracy

# Modeling - Simple Manual Data
Let's manually look at our data and decide some cutoff points for classification

### Decsion Tree Model

```python
data["target_names"]
```

```python
def single_feature_prediction(petal_length):
    """ Predicts the iris specied on the basis of peta length"""
    if petal_length < 2.5:
        return 0
    elif petal_length < 4.8:
        return 1
    else: 
        return 2
```

```python
# df_train.columns                    # to locate petal_length column
print(x_train[:, 2])                    # all rows and 2-indexed column (for petal_length)
```

```python
mannual_predictions = np.array([single_feature_prediction(val) for val in x_train[:, 2]])
```

```python
print(mannual_predictions == y_train)         # coz we are matching our prediction to y_train
```

for what percent or prediction/classification is correct

```python
mannual_model_accuracy = np.mean(mannual_predictions==y_train)

print(f"Manuual Model accuracy to classify the iris specie is : {mannual_model_accuracy * 100:.2f}%")
```

# Modeling - Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter=200)
```

```python
# tarining our model using .fit() method
model.fit(x_train,y_train)
```

### Prediciting the Model Classification

```python
y_pred=model.predict(x_train)          # return the array of 0,1 2 of species targets
```

### Model Accuracy
Both below .mean() and .score() methods are used for accuracies

```python
np.mean(y_pred == y_train)
```

```python
final = model.score(x_train,y_train)
```

```python
print(f"Logistic Model accuracy to classify the iris specie is : {final * 100:.2f}%")
```
